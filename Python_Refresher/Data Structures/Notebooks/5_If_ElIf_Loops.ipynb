```python
def categorize_age(age):
    if age < 13:
        return "Child"
    elif 13 <= age < 18:
        return "Teen"
    else:
        return "Adult"
```

```python
age_category = categorize_age(16)
print(age_category)
```

```python
age_category = categorize_age(11)
print(age_category)
```

```python
age_category = categorize_age(30)
print(age_category)
```

```python
stored_inputs = []
```

```python
user_input = input("Enter a value (type 'exit' to stop): ")
```

```python
while user_input.lower() != "exit":
    user_input = input("Enter a value (type 'exit' to stop): ")
    stored_inputs.append(user_input)
```

```python
# Output the stored inputs
print("Stored inputs:", stored_inputs)
```

```python
break_example = []
for num in range(1, 10):
    if num == 5:
        break  # Stops at 5
    break_example.append(num)
print(break_example)
```

```python
continue_example = []
for num in range(1, 10):
    if num == 5:
        continue  # Skips 5
    continue_example.append(num)

print(continue_example)
```
