```python
def greet(name, greeting="Hello"):
    """Function to greet a person with a custom message."""
    return f"{greeting}, {name}!"
```

```python
name_arg = "Alice"
greeting_arg = "Good morning"
# (Positional argument matching)
result1 = greet(name_arg, greeting_arg)

# (Keyword argument matching)
result2 = greet(name=name_arg, greeting=greeting_arg)

# (Default argument usage)
result3 = greet(name_arg)  # Uses default greeting
```

```python
print(result1) 
print(result2) 
print(result3) 
```
