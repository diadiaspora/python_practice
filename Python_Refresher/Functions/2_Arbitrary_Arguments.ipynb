```python
def display_items(*args, **kwargs):
    """Function to accept and display arbitrary positional and keyword arguments."""

    # Display positional arguments
    print("Positional Arguments:")
    for index, item in enumerate(args, start=1):
        print(f"  Item {index}: {item}")

    # Display keyword arguments
    print("\nKeyword Arguments:")
    for key, value in kwargs.items():
        print(f"  {key}: {value}")
```

how do we know one is for a tuple and one is for a dictionary?

Short answer: because of Python‚Äôs syntax rules, not because of the variable names.

Here‚Äôs the clean explanation.

1. It‚Äôs the * vs **, not args vs kwargs

Python decides what container to use based on the number of asterisks:

Syntax	Meaning	Data type created
*args	Collect extra positional arguments	tuple
**kwargs	Collect extra keyword arguments	dict

The names args and kwargs are just conventions.
You could write this (bad style, but valid):

def f(*bananas, **apples):
    print(type(bananas))
    print(type(apples))


Output will always be:

<class 'tuple'>
<class 'dict'>

2. Why * creates a tuple

When you pass positional arguments:

f(1, 2, 3)


Python has no keys, only values, so it stores them in an ordered, immutable sequence ‚Üí a tuple.

That‚Äôs why:

*args  ‚Üí (1, 2, 3)

3. Why ** creates a dictionary

Keyword arguments are passed as:

f(a=1, b=2)


Each value has a name (key), so Python must store them as:

**kwargs ‚Üí {'a': 1, 'b': 2}


Keys are required ‚Üí dictionary is the only logical structure.

4. You can prove it yourself
def test(*x, **y):
    print(type(x))
    print(type(y))

test(10, 20, name="Dia", age=30)


Output:

<class 'tuple'>
<class 'dict'>


This is guaranteed by the Python language spec.

5. Mental model (easy to remember)

* ‚Üí ‚Äúmany values‚Äù ‚Üí tuple

** ‚Üí ‚Äúmany named values‚Äù ‚Üí dictionary

Bottom line

You don‚Äôt infer it ‚Äî Python enforces it:

One * always packs positional arguments into a tuple

Two ** always pack keyword arguments into a dictionary

so what is a positional argument and what is a keyword argument?

Here‚Äôs the clean, no-fluff explanation, starting from first principles.

1. Positional arguments

A positional argument is passed to a function based on its position (order).

Example
def greet(name, age):
    print(name, age)

greet("Dia", 30)


Python assigns:

"Dia" ‚Üí name (1st position)

30 ‚Üí age (2nd position)

If you swap them:

greet(30, "Dia")


Python does not know your intent ‚Äî it follows position:

30 Dia

Key rule

Positional arguments are matched by order, not by name.

2. Keyword arguments

A keyword argument is passed using the parameter name.

Example
greet(age=30, name="Dia")


Now:

Order does not matter

Each value is explicitly tied to a name

This works:

greet(name="Dia", age=30)
greet(age=30, name="Dia")

Key rule

Keyword arguments are matched by name, not position.

3. Same function, both types
greet("Dia", age=30)


"Dia" ‚Üí positional

age=30 ‚Üí keyword

This is very common.

4. The strict ordering rule

Positional arguments must come before keyword arguments.

‚ùå Invalid:

greet(age=30, "Dia")


‚úÖ Valid:

greet("Dia", age=30)


Reason: once Python sees a keyword, it can‚Äôt guess which remaining positions you meant.

5. How this relates to *args and **kwargs
def show(*args, **kwargs):
    print(args)
    print(kwargs)


Call:

show(1, 2, x=10, y=20)


Python does this:

Positional arguments ‚Üí packed into args (tuple)

Keyword arguments ‚Üí packed into kwargs (dict)

Result:

args   = (1, 2)
kwargs = {'x': 10, 'y': 20}

6. Defaults blur the line (important detail)
def greet(name, age=30):
    print(name, age)


You can still pass age positionally:

greet("Dia", 25)


Or as a keyword:

But position always wins first.

7. Mental model (remember this)

Positional ‚Üí ‚Äúfirst goes here, second goes there‚Äù

Keyword ‚Üí ‚Äúthis value belongs to this name‚Äù

One-sentence summary

Positional arguments are assigned by order; keyword arguments are assigned by name.

What Python actually cares about

Python only cares about the asterisks:

* ‚Üí collect positional arguments into a tuple

** ‚Üí collect keyword arguments into a dictionary

The variable names themselves are completely up to you.
